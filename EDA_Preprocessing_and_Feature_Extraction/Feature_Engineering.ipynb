```python
# import modules
import pandas as pd
from collections import Counter
```

```python
# create dataframe
df1 = pd.read_csv('data_prml.csv')
df1
```

```python
# For Title and tmdbId
df1.dropna(inplace=True)

# For genre, Director, Stars and Keywords
df1 = df1[df1['genres'] != "['(no genres listed)']"]
df1 = df1[df1['Director'] != "['']"]
df1 = df1[df1['Stars'] != "['']"]
df1 = df1[df1['Keywords'] != "['']"]

# Year
all_years = []
for i in df1['Year']:
    all_years.append(i)

# Corresponding counts and filtering
counts = Counter(all_years)
# Get years in which atleast 10 movies have released
filtered_counts = {key: value for key, value in counts.items() if value >= 10}
filtered_keys = filtered_counts.keys()
indii = []
for i in range(len(all_years)):
    if all_years[i] in filtered_keys:
        indii.append(i)
df1 = df1.iloc[indii]
```

```python
# Dropping unwanted columns
df1.drop(columns = ['imdbId', 'tmdbId', 'Keywords', 'Title'], inplace = True)
df1
```

* Converting columns to usable type

```python
# genre
Genre = []
for i in df1['genres']:
    i = i.replace('[', '').replace(']', '').replace('"', '').replace("'", '').replace(' ', '')
    i = i.lower()
    Genre.append(i.split(','))
df1['genres'] = Genre
```

```python
# Director
dir = []
for i in df1['Director']:
    i = i.replace('[', '').replace(']', '').replace('"', '').replace("'", '')
    dd = ''
    le = len(i.split(', '))
    for j in i.split(', '):
        le -= 1
        j = j.replace(' ', '')
        j = j.lower()
        dd += j
        if(le != 0):
            dd += ' '
    dir.append(dd.split(' '))

# Get directors who have atleast directed 5 movies
all_dir = []
for i in dir:
    for j in i:
        all_dir.append(j)
# Corresponding counts and filtering
counts = Counter(all_dir)
filtered_counts = {key: value for key, value in counts.items() if value >= 5}

# Create a new_list and make sure to remove no director movies
top_dir = []
for i in dir:
    d = []
    for j in i:
        if j in all_dir:
            d.append(j)
    top_dir.append(d)
df1['Director'] = top_dir
```

```python
# Stars
sta = []
for i in df1['Stars']:
    i = i.replace('[', '').replace(']', '').replace('"', '').replace("'", '')
    dd = ''
    le = len(i.split(', '))
    for j in i.split(', '):
        le -= 1
        j = j.replace(' ', '')
        j = j.lower()
        dd += j
        if(le != 0):
            dd += ' '
    sta.append(dd.split(' '))
    
# Get stars who have atleast acted in 5 movies
all_stars = []
for i in sta:
    for j in i:
        all_stars.append(j)
# Corresponding counts and filtering
counts = Counter(all_stars)
filtered_counts = {key: value for key, value in counts.items() if value >= 5}

# Create a new_list of top_stars
top_stars = []
for i in sta:
    s = []
    for j in i:
        if j in all_stars:
            s.append(j)
    top_stars.append(s)
df1['Stars'] = top_stars 
```

```python
# Create an empty data_frame 
df = pd.DataFrame()
```

```python
# We are going create multiple instances of same movies
# as there are lists of genre, directors and stars
for index, row in df1.iterrows():
    r = []
    # movieId
    r.append(row[0])
    for g in row[1]:
        # genre
        r.append(g)
        # Year
        r.append(row[2])
        for d in row[3]:
            # director
            r.append(d)
            for s in row[4]:
                # star
                r.append(s)
                # add row
                df = pd.concat([df, pd.DataFrame([r])], ignore_index=True)
                r.pop()
            r.pop()
        r.pop()
        r.pop()
```

```python
# edit column names
df.columns = df1.columns
df
```

```python
# keep count of how many instaces you have created as it will
# help later during look_up tables
movie_counts = df['movieId'].value_counts()
movie_counts
```
